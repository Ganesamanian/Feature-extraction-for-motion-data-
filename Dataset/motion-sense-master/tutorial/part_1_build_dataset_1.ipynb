### Data Info
(Source: https://bmi.teicrete.gr/en/the-mobifall-and-mobiact-datasets-2/) 
- 4 different types of `falls` performed by 66 participants
- 11 different types of `ADLs` performed by 19 participants and 9 types of ADLs performed by 59 participants (plus one activity `LYI` which results from the `inactivity period after a fall` by 66 participants)
- Five sub-scenarios which construct one scenario of daily living, which consists of a sequence of 50 activities and performed by 19 participants.


- The annotated data in csv format, separated by each activity

#### File names:
Filename format:
```
<ADL OR FALL OR SCENARIO_CODE>_<SUBJECT_ID>_<TRIAL_NO>.txt
```

````python
import numpy as np
import os
import pandas as pd
ds_inf = pd.read_csv('mobiAct_ds_inf.txt', sep="\s*[|]\s*", engine='python')
ds_inf = ds_inf.drop(columns=['Unnamed: 0', 'Name', 'Surname', 'Unnamed: 8'], axis=1)
ds_inf
````

````python
dir_name = "Annotated Data" 
act_folders = [x for x in os.listdir(dir_name) if len(x)==3]
act_folders
````

````python
columns = [#'timestamp','rel_time',
                'acc_x','acc_y','acc_z',
                'gyro_x', 'gyro_y', 'gyro_z',
                'azimuth','pitch','roll','label']
columns.extend(['act','age', 'height', 'weight','gender', 'pid', 'trial'])
columns
````

````python
train_dataset = pd.DataFrame(columns=columns)
test_dataset = pd.DataFrame(columns=columns)
train_test_ratio = 0.67
for i, act in enumerate(act_folders):
    act_dirs = os.listdir(dir_name+"/"+act)
    for act_file in act_dirs:
        print("***",act,round(i/len(act_folders), 2)*100,"%", 
              act_file, "train: ",len(train_dataset), "test: ", len(test_dataset), end = "\r")
        # <ACT_CODE>_<SUBJECT_ID>_<TRIAL_NO>        
        pid = int(act_file.split('_')[1])
        trial = int(act_file.split('_')[2])
        tmp_inf = ds_inf[ds_inf["ID"]==pid]
        p_gender = str(tmp_inf.iloc[0]["Gender"]).strip()
        p_age = int(tmp_inf.iloc[0]["Age"])
        p_height = int(tmp_inf.iloc[0]["Height"])
        p_weight = int(tmp_inf.iloc[0]["Weight"])
        
        tmp_data = pd.read_csv(dir_name+"/"+act+"/"+act_file)
        tmp_data = tmp_data.drop(columns=['timestamp','rel_time'], axis=1)
        
        tmp_data['gender'] = p_gender
        tmp_data['age'] = p_age
        tmp_data['height'] = p_height
        tmp_data['weight'] = p_weight
        tmp_data['pid'] = pid-1
        tmp_data['trial'] = trial
        tmp_data['act'] = act
        
        if act in ['STD', 'WAL', 'SIT']:
            tmp_size = len(tmp_data)
            train_dataset = train_dataset.append(tmp_data[:int(train_test_ratio*tmp_size)], ignore_index=True, sort=False)
            test_dataset = test_dataset.append(tmp_data[int(train_test_ratio*tmp_size):], ignore_index=True, sort=False)
        else:
            if trial%3 != 0:
                train_dataset = train_dataset.append(tmp_data, ignore_index=True, sort=False)
            else:
                test_dataset = test_dataset.append(tmp_data, ignore_index=True, sort=False)
    print("\n",act,round(i/len(act_folders), 2)*100,"%","\n")
train_dataset.shape, test_dataset.shape
````
